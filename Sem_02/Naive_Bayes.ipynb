````python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris # Iris dataset 
````

````python
iris = load_iris()  # Cargar el dataset Iris
iris_x = iris.data[:, [0, 2]]  # Seleccionamos las características 0 y 2
iris_y = iris.target  # 

# Imprimir las formas (dimensiones) de las matrices
print(iris_x.shape, iris_y.shape)
````

````python
iris = load_iris()
````

````python
# Instance kNN object
naive_bayes = GaussianNB()

# Training
naive_bayes.fit( iris_x , iris_y )
````

````python
## Plot decision boundaries
#______________________________________
# Definir los límites de la malla
x_min, x_max = iris_x[:, 0].min() - 1, iris_x[:, 0].max() + 1
y_min, y_max = iris_x[:, 1].min() - 1, iris_x[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),# Creamos la malla de puntos
                     np.arange(y_min, y_max, 0.1))
Z = naive_bayes.predict(np.c_[xx.ravel(), yy.ravel()])# Predecir las clases para cada punto en la malla
Z = Z.reshape(xx.shape)# Cambiar la forma de Z para que coincida con la malla                     
````

## On real data - Fruits dataset

Download the fruits data [here](https://drive.google.com/file/d/1M8tiAWDZclABJN1Meq9oEHCLJTNyHdpP/view?usp=share_link)

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
````

````python
from sklearn.datasets import load_iris  # esta seria  la función que cargamos mi dataset de frutas    # preguntar profe por el dataset _ pero Supongamos que tienes un dataset llamado "fruits"
import numpy as np
# Load the fruits dataset
fruit_data = load_iris()  # aca cargamos el data de la dataset del profe

# Selecciona 2 características (e.g., características 0 y 2 como en el dataset Iris)
fruits_x = fruit_data.data[:, [0, 2]]  # Seleccionamos las columnas 0 y 2
fruits_y = fruit_data.target  # Etiquetas (clases)

# Convertir todo a un array de NumPy
fruits_x, fruits_y = np.asarray(fruits_x), np.asarray(fruits_y)

# Imprimir las dimensiones
print(fruits_x.shape, fruits_y.shape)
````

````python
# Instance GaussianNB object
naive_bayes = GaussianNB()

# Training
naive_bayes.fit(fruits_x, fruits_y)  # (num_samples, Features)
````

````python
import numpy as np
import matplotlib.pyplot as plt


x_min, x_max = fruits_x[:, 0].min() - 1, fruits_x[:, 0].max() + 1# Definir los límites de la malla basados en los datos
y_min, y_max = fruits_x[:, 1].min() - 1, fruits_x[:, 1].max() + 1


xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),# Crear la malla de puntos
                     np.arange(y_min, y_max, 0.1))

Z = naive_bayes.predict(np.c_[xx.ravel(), yy.ravel()])# Predecir las clases para cada punto en la malla
Z = Z.reshape(xx.shape)# Cambiar la forma de Z para que coincida con la malla
plt.figure(figsize=(8, 6))# Crear el gráfico
plt.contourf(xx, yy, Z, alpha=0.5, cmap='viridis')#frontesras de deciciones
plt.scatter(fruits_x[:, 0], fruits_x[:, 1], c=fruits_y, cmap='viridis', edgecolors='k')# Trazamos los puntos de datos originales

# Etiquetas y título
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Fronteras de Decisión con Naive Bayes')

# Mostrar el gráfico
plt.show()
````
